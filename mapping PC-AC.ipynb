```python
import os
import openpyxl

# Folder path containing Excel files
folder_path = '/Users/apple/Desktop/PC-AC list from LGD 2'

# Iterate over all files in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.xlsx'):
        file_path = os.path.join(folder_path, file_name)

        # Open the Excel file
        workbook = openpyxl.load_workbook(file_path)

        # Iterate over all sheets in the workbook
        for sheet_name in workbook.sheetnames:
            sheet = workbook[sheet_name]

            # Iterate over all cells in the sheet
            for row in sheet.iter_rows():
                for cell in row:
                    # Check if the cell contains text
                    if cell.data_type == 's' and cell.value is not None:
                        cell.value = str(cell.value) + ','  # Add a comma at the end of the text

        # Save the modified Excel file
        output_file_path = os.path.join(folder_path, f"modified_{file_name}")
        workbook.save(output_file_path)

        print(f"Comma added at the end of every cell for file: {file_name}")

print("Modification completed for all Excel files in the folder.")
```

```python
import os
import pandas as pd

# Function to recursively search for Excel files
def search_excel_files(folder_path):
    excel_files = []
    if os.path.exists(folder_path):
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                if file.lower().replace(" ", "") == "Mankachar.xlsx":
                    excel_files.append(os.path.join(root, file))
    return excel_files

# Set up the folder paths
assam_folder = '/Users/apple/Desktop/assam'
output_filename = '/Users/apple/Desktop/Combined_Mankachar.xlsx'

# Create an empty DataFrame to store the combined data
combined_data = pd.DataFrame()

# Iterate over the "assam" folders and their subfolders
for root, dirs, files in os.walk(assam_folder):
    for dir_name in dirs:
        dhubri_folder_path = os.path.join(root, dir_name, 'Dhubri')
        if os.path.exists(dhubri_folder_path):
            excel_files = search_excel_files(dhubri_folder_path)
            for file_path in excel_files:
                df = pd.read_excel(file_path)
                combined_data = pd.concat([combined_data, df], ignore_index=True)

# Save the combined DataFrame to a new Excel file
combined_data.to_excel(output_filename, index=False)
```

```python
import os
import shutil

# Function to create a new folder
def create_folder(folder_path):
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)

# Function to move file to a folder
def move_file(file_path, folder_path):
    shutil.move(file_path, folder_path)

# Function to remove digits from file name
def remove_digits(file_name):
    return ''.join(filter(lambda x: not x.isdigit(), file_name))

# Function to convert file name to lowercase
def convert_to_lowercase(file_name):
    return file_name.lower()

# Set up the main folder path
main_folder = '/Users/apple/Desktop/arunachal pradesh 2019'
output_folder = '/Users/apple/Desktop/arunachal pradesh 2019 try'

# Iterate over the subfolders within the main folder
for root, dirs, files in os.walk(main_folder):
    for dir_name in dirs:
        folder_path = os.path.join(root, dir_name)

        # Iterate over the Excel files in the subfolder
        for filename in os.listdir(folder_path):
            if filename.lower().endswith('.xlsx'):
                file_name = os.path.splitext(filename)[0]
                file_name = remove_digits(file_name)
                file_name = convert_to_lowercase(file_name)

                # Create the corresponding folder in the output folder
                output_subfolder = os.path.join(output_folder, file_name)
                create_folder(output_subfolder)

                # Move the file to the output subfolder
                file_path = os.path.join(folder_path, filename)
                move_file(file_path, output_subfolder)
                
```

```python
import os
import shutil

# Define the paths to the folders
folder_2013_path = '/Users/apple/Desktop/tripura/tripura 2013'
folder_2018_path = '/Users/apple/Desktop/tripura/tripura 2018'
combined_folder_path = '/Users/apple/Desktop/combined_tripura'


# Create the combined folder if it doesn't exist
os.makedirs(combined_folder_path, exist_ok=True)

# Function to combine Excel files with the same name
def combine_excel_files(folder1, folder2, combined_folder):
    # Iterate over the subfolders in the first folder
    for subfolder in os.listdir(folder1):
        subfolder_path = os.path.join(folder1, subfolder)
        if os.path.isdir(subfolder_path):
            # Iterate over the files in the subfolder
            for file in os.listdir(subfolder_path):
                if file.endswith('.xlsx'):
                    file_path = os.path.join(subfolder_path, file)
                    file_name, _ = os.path.splitext(file)

                    # Check if the corresponding file exists in the second folder
                    matching_subfolder_path = os.path.join(folder2, subfolder)
                    matching_file_path = os.path.join(matching_subfolder_path, file)
                    if os.path.isfile(matching_file_path):
                        # Combine the two files into a single folder
                        combined_file_path = os.path.join(combined_folder, file_name, file)
                        os.makedirs(os.path.dirname(combined_file_path), exist_ok=True)
                        shutil.copy(file_path, combined_file_path)
                        shutil.copy(matching_file_path, combined_file_path)

# Combine the Excel files based on the same names
combine_excel_files(folder_2013_path, folder_2018_path, combined_folder_path)

print('Excel files combined successfully.')
```

```python
import os
import shutil
import openpyxl

# Define the paths to the folders
folder_2013_path = '/Users/apple/Desktop/uttar pradesh/uttar pradesh 2012'
folder_2018_path = '/Users/apple/Desktop/uttar pradesh/uttar pradesh 2017'
folder_2022_path = '/Users/apple/Desktop/uttar pradesh/uttar pradesh 2022'
combined_folder_path = '/Users/apple/Desktop/uttar pradesh_comb'

# Create the combined folder if it doesn't exist
os.makedirs(combined_folder_path, exist_ok=True)

# Function to combine Excel files with the same name
def combine_excel_files(folder1, folder2, folder3, combined_folder):
    # Iterate over the subfolders in the first folder
    for subfolder in os.listdir(folder1):
        subfolder_path = os.path.join(folder1, subfolder)
        if os.path.isdir(subfolder_path):
            # Get the corresponding subfolders in the other folders
            matching_subfolder_path1 = os.path.join(folder1, subfolder)
            matching_subfolder_path2 = os.path.join(folder2, subfolder)
            matching_subfolder_path3 = os.path.join(folder3, subfolder)
            if (
                os.path.isdir(matching_subfolder_path1) and
                os.path.isdir(matching_subfolder_path2) and
                os.path.isdir(matching_subfolder_path3)
            ):
                # Iterate over the files in the subfolder

                for file in os.listdir(subfolder_path):
                    if file.endswith('.xlsx'):
                        file_name, _ = os.path.splitext(file)
                        file_path1 = os.path.join(matching_subfolder_path1, file)
                        file_path2 = os.path.join(matching_subfolder_path2, file)
                        file_path3 = os.path.join(matching_subfolder_path3, file)
                        if (
                            os.path.isfile(file_path1) and
                            os.path.isfile(file_path2) and
                            os.path.isfile(file_path3)
                        ):
                            # Combine the files into a single file
                            combined_file_path = os.path.join(combined_folder, f'{file_name}.xlsx')

                            # Open the files
                            workbook1 = openpyxl.load_workbook(file_path1)
                            sheet_names = workbook1.sheetnames
                            sheet1 = workbook1[sheet_names[0]]

                            workbook2 = openpyxl.load_workbook(file_path2)
                            sheet2 = workbook2[sheet_names[0]]

                            workbook3 = openpyxl.load_workbook(file_path3)
                            sheet3 = workbook3[sheet_names[0]]

                            # Combine the data from all files into a new file
                            combined_workbook = openpyxl.Workbook()
                            combined_sheet = combined_workbook.active

                            # Copy data from the first file
                            for row in sheet1.iter_rows(values_only=True):
                                combined_sheet.append(row)

                            # Copy data from the second file, skipping the header row
                            for row in sheet2.iter_rows(min_row=2, values_only=True):
                                combined_sheet.append(row)

                            # Copy data from the third file, skipping the header row
                            for row in sheet3.iter_rows(min_row=2, values_only=True):
                                combined_sheet.append(row)

                            # Save the combined file
                            combined_workbook.save(combined_file_path)

# Combine the Excel files based on the same names
combine_excel_files(folder_2013_path, folder_2018_path, folder_2022_path, combined_folder_path)

print('Excel files combined successfully.')
```

```python
import os
import shutil
import openpyxl

# Define the paths to the folders
folder_2012_path = '/Users/apple/Desktop/2 folders/tripura/tripura 2013'
folder_2017_path = '/Users/apple/Desktop/2 folders/tripura/tripura 2018'
folder_2022_path = '/Users/apple/Desktop/2 folders/tripura/del'
combined_folder_path = '/Users/apple/Desktop/combined_path/tripura'


# Create the combined folder if it doesn't exist
os.makedirs(combined_folder_path, exist_ok=True)

# Function to combine Excel files with the same name
def combine_excel_files(folder1, folder2, folder3, combined_folder):
    # Get the list of all subfolders in the first folder
    subfolders = set(os.listdir(folder1))
    subfolders |= set(os.listdir(folder2))
    subfolders |= set(os.listdir(folder3))

    # Iterate over the subfolders
    for subfolder in subfolders:
        subfolder_path1 = os.path.join(folder1, subfolder)
        subfolder_path2 = os.path.join(folder2, subfolder)
        subfolder_path3 = os.path.join(folder3, subfolder)

        # Create subfolder in the combined folder
        combined_subfolder_path = os.path.join(combined_folder, subfolder)
        os.makedirs(combined_subfolder_path, exist_ok=True)

        # Combine the files into a single file within the subfolder
        for folder_path, folder_num in [(subfolder_path1, 1), (subfolder_path2, 2), (subfolder_path3, 3)]:
            if os.path.isdir(folder_path):
                for file in os.listdir(folder_path):
                    if file.endswith('.xlsx'):
                        file_name, _ = os.path.splitext(file)
                        file_path = os.path.join(folder_path, file)
                        combined_file_path = os.path.join(combined_subfolder_path, f'{file_name}.xlsx')

                        if os.path.isfile(file_path):
                            # Open the file
                            workbook = openpyxl.load_workbook(file_path)
                            sheet_names = workbook.sheetnames
                            sheet = workbook[sheet_names[0]]

                            if os.path.isfile(combined_file_path):
                                # Append data to the existing combined file
                                combined_workbook = openpyxl.load_workbook(combined_file_path)
                                combined_sheet = combined_workbook.active

                                # Copy data from the file, including the header row
                                for row in sheet.iter_rows(values_only=True):
                                    combined_sheet.append(row)

                                # Save the combined file
                                combined_workbook.save(combined_file_path)

                                print(f'File appended: {file_name}.xlsx')
                            else:
                                # Create a new combined file and copy data from the file
                                shutil.copyfile(file_path, combined_file_path)

                                print(f'File added: {file_name}.xlsx')

                        else:
                            # Create an empty file in the combined folder
                            combined_file_path = os.path.join(combined_subfolder_path, f'{file_name}.xlsx')
                            open(combined_file_path, 'w').close()

                            print(f'File added: {file_name}.xlsx (No match in folder {folder_num})')

# Combine the Excel files based on the same names
combine_excel_files(folder_2012_path, folder_2017_path, folder_2022_path, combined_folder_path)

print('Excel files combined successfully.')
```

```python
import os
import openpyxl
from openpyxl.styles import Font, PatternFill, Alignment 

# Specify the folder path where the Excel files are located
folder_path = '/Users/apple/Desktop/nagaland 2018'

# Define the formatting options
highlight_color = 'FFFF00'  # Yellow color for highlighting
font_size = 16  # Font size for the added row


# Iterate over the subfolders in the main folder
for subfolder in os.listdir(folder_path):
    subfolder_path = os.path.join(folder_path, subfolder)
    if os.path.isdir(subfolder_path):
        # Extract the year from the subfolder name
        year = subfolder.split()[-1]

        # Iterate over the sub-subfolders in the subfolder
        for sub_subfolder in os.listdir(subfolder_path):
            sub_subfolder_path = os.path.join(subfolder_path, sub_subfolder)
            if os.path.isdir(sub_subfolder_path):
                # Iterate over the files in the sub-subfolder
                for file_name in os.listdir(sub_subfolder_path):
                    if file_name.endswith('.xlsx') or file_name.endswith('.xls'):
                        file_path = os.path.join(sub_subfolder_path, file_name)

                        # Open the Excel file
                        workbook = openpyxl.load_workbook(file_path)
                        sheet = workbook.active

                        # Insert the year value in the first row of the sheet
                        sheet.insert_rows(1)
                        sheet.cell(row=1, column=1).value = year

                        # Apply formatting to the added row
                        added_row = sheet[1]
                        for cell in added_row:
                            cell.font = Font(size=font_size)
                            cell.fill = PatternFill(fill_type='solid', fgColor=highlight_color)
                            cell.alignment = Alignment(vertical='center', horizontal='center')

                        # Merge the cells in the added row
                        sheet.merge_cells(start_row=1, start_column=1, end_row=1, end_column=sheet.max_column)

                        # Save the modified file
                        workbook.save(file_path)

                        print(f'Year "{year}" added and formatted in file "{file_name}"')
```

```python
import os
import pandas as pd

def remove_and_rename_columns(file_path):
    # Read the Excel file, skipping the first row
    df = pd.read_excel(file_path, skiprows=1)
    
    # Remove leading and trailing spaces from column names
    df.columns = df.columns.str.strip()
    
    # Check if the columns exist before removing them
    columns_to_remove = ["STATE/UT NAME", "AC NO.","Constituency No."]
    existing_columns = df.columns.tolist()
    columns_to_remove_existing = [col for col in columns_to_remove if col in existing_columns]
    
    if columns_to_remove_existing:
        df.drop(columns_to_remove_existing, axis=1, inplace=True)
    
    # Rename columns
    df.rename(columns={
        "AC NAME": "Constituency Name",
    }, inplace=True)
    
    # Save the modified DataFrame back to the Excel file
    df.to_excel(file_path, index=False)

# Provide the directory path
directory_path = '/Users/apple/Desktop/meghalaya 2018'

# Traverse through the directory and its subfolders
for root, dirs, files in os.walk(directory_path):
    for file in files:
        if file.endswith('.xlsx'):
            file_path = os.path.join(root, file)
            
            # Remove and rename columns in each Excel file
            remove_and_rename_columns(file_path)

print("Columns removed and renamed successfully.")
```

```python
import os
import openpyxl
from openpyxl.styles import Font, PatternFill, Alignment

# Specify the folder path where the Excel files are located
folder_path = '/Users/apple/Desktop/nagaland 2018'

# Define the formatting options
highlight_color = 'FFFF00'  # Yellow color for highlighting
font_size = 16  # Font size for the added row

# Iterate over the subfolders in the main folder
for subfolder in os.listdir(folder_path):
    subfolder_path = os.path.join(folder_path, subfolder)
    if os.path.isdir(subfolder_path):
        # Extract the year from the subfolder name
        year = subfolder.split()[-1]

        # Iterate over the sub-subfolders in the subfolder
        for sub_subfolder in os.listdir(subfolder_path):
            sub_subfolder_path = os.path.join(subfolder_path, sub_subfolder)
            if os.path.isdir(sub_subfolder_path):
                # Iterate over the files in the sub-subfolder
                for file_name in os.listdir(sub_subfolder_path):
                    if file_name.endswith('.xlsx') or file_name.endswith('.xls'):
                        file_path = os.path.join(sub_subfolder_path, file_name)

                        # Open the Excel file
                        workbook = openpyxl.load_workbook(file_path)
                        sheet = workbook.active

                        # Insert the year value in the first row of the sheet
                        sheet.insert_rows(1)
                        sheet.cell(row=1, column=1).value = year

                        # Apply formatting to the added row
                        added_row = sheet[1]
                        for cell in added_row:
                            cell.font = Font(size=font_size)
                            cell.fill = PatternFill(fill_type='solid', fgColor=highlight_color)
                            cell.alignment = Alignment(vertical='center', horizontal='center')

                        # Merge the cells in the added row
                        sheet.merge_cells(start_row=1, start_column=1, end_row=1, end_column=sheet.max_column)

                        # Find the last row with data in the sheet
                        last_row = sheet.max_row

                        # Iterate over the rows in reverse order starting from the last row
                        for row in range(last_row, 1, -1):
                            # Check if all cells in the row are empty
                            if all(not sheet.cell(row=row, column=col).value for col in range(1, sheet.max_column + 1)):
                                # Delete the empty row
                                sheet.delete_rows(row)

                        # Save the modified file
                        workbook.save(file_path)

                        print(f'Year "{year}" added and formatted in file "{file_name}"')
```

### adds one row and merges over rows and adds year and assembly elections to it

```python
import os
import shutil
import openpyxl

# Define the path to the main folder
main_folder_path = '/Users/apple/Desktop/meghalaya'

# Function to add '(ASSEMBLY ELECTIONS)' after the year in each file
def add_assembly_elections(folder_path):
    # Extract the year from the root folder's name
    year = os.path.basename(folder_path).split()[1]

    # Iterate over the subfolders in the main folder
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith('.xlsx'):
                file_path = os.path.join(root, file)

                # Open the file
                workbook = openpyxl.load_workbook(file_path)
                sheet_names = workbook.sheetnames
                sheet = workbook[sheet_names[0]]

                # Insert a new row at the beginning
                sheet.insert_rows(1)

                # Merge the new row with existing cells
                sheet.merge_cells(start_row=1, start_column=1, end_row=1, end_column=sheet.max_column)

                # Get the first cell in the new row and set its value as the modified year
                first_cell = sheet.cell(row=1, column=1)
                first_cell.value = f'{year} (ASSEMBLY ELECTIONS)'

                # Save the modified file
                workbook.save(file_path)

                print(f'File modified: {file}')

# Add '(ASSEMBLY ELECTIONS)' after the year in each file within the main folder
add_assembly_elections(main_folder_path)

print('Modification completed successfully.')
```
