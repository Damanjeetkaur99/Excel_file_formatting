#### use this for renaming files in a file

```python
import os
import pandas as pd
import warnings
from openpyxl import load_workbook

warnings.filterwarnings('ignore')

folder_path = "/Users/apple/Desktop/arunachal pradesh 2009 form 20 xl"  # Replace with the actual folder path folder

# Iterate through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".xlsx") or filename.endswith(".xls"):  # Process only Excel files
        file_path = os.path.join(folder_path, filename)
        try:
            # Load the workbook using openpyxl
            wb = load_workbook(file_path, data_only=True)

            # Get the first worksheet
            sheet = wb.worksheets[0]

            # Get the value from the merged cell in the first row and first column
            cell_value = sheet.cell(row=1, column=1).value

            # Convert the value to a string and remove any leading/trailing whitespaces and special characters
            cell_value = str(cell_value).strip()

            # Remove any invalid characters from the cell value that are not allowed in filenames
            invalid_chars = r'\/:*?"<>|'
            for char in invalid_chars:
                cell_value = cell_value.replace(char, "")
                

            # Rename the file using the extracted text
            new_filename = cell_value + ".xlsx"  # You can adjust the file extension if needed

            # Rename the file
            new_file_path = os.path.join(folder_path, new_filename)
            os.rename(file_path, new_file_path)

            print(f"Renamed '{filename}' to '{new_filename}'")
        except Exception as e:
            print(f"Error processing file '{filename}': {str(e)}")
```

```python
import os
import pandas as pd
import warnings
import re
warnings.filterwarnings('ignore')

folder_path = "/Users/apple/Desktop/delhi 2013 form 20 xl "  # Replace with the actual folder path folder

# Regular expression pattern to match the required text pattern
pattern = r"Name of assembly(.+)"

# Iterate through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".xlsx") or filename.endswith(".xls"):  # Process only Excel files
        file_path = os.path.join(folder_path, filename)
        try:
            # Load the workbook using pandas
            df = pd.read_excel(file_path, header=None)

            # Search for the required text pattern across all rows
            constituency_name = None
            for row in df.iterrows():
                for cell_value in row[1]:
                    cell_value = str(cell_value).strip()  # Remove leading and trailing spaces
                    match = re.search(pattern, cell_value)
                    if match:
                        constituency_name = match.group(1).strip()
                        break

                if constituency_name:
                    break

            if constituency_name:
                # Remove any invalid characters from the cell value that are not allowed in filenames
                
                invalid_chars = r'\/:*?"<>|'   
                for char in invalid_chars:
                    constituency_name = constituency_name.replace(char, "") 

                # Rename the file using the extracted text
                new_filename = constituency_name + ".xlsx"  # You can adjust the file extension if needed 
    
                # Rename the file
                new_file_path = os.path.join(folder_path, new_filename)
                os.rename(file_path, new_file_path)
     

                print(f"Renamed '{filename}' to '{new_filename}'")
        
            else:
                print(f"File '{filename}' does not contain the required text pattern.")
        except Exception as e:
            print(f"Error processing file '{filename}': {str(e)}")
```

```python
import os

# Specify the folder path where the Excel files are located
folder_path = '/Users/apple/Desktop/arunachal pradesh 2019 form 20 xl'

# Specify the text you want to remove from the file names
text_to_remove = ('/')

# Get a list of all files in the folder
files = os.listdir(folder_path)

# Iterate over the files in the folder
for file_name in files:
    if file_name.endswith('.xlsx') or file_name.endswith('.xls'):
        # Generate the new file name by replacing the specific text with an empty string
        new_file_name = file_name.replace(text_to_remove, '')

        # Construct the full paths of the old and new file names
        old_file_path = os.path.join(folder_path, file_name)
        new_file_path = os.path.join(folder_path, new_file_name)

        # Rename the file
        os.rename(old_file_path, new_file_path)
        print(f'Renamed "{file_name}" to "{new_file_name}"')
```

```python
import os
import re

# Specify the folder path where the Excel files are located
folder_path = '/Users/apple/Desktop/arunachal pradesh 2019 form 20 xl'

# Define a regular expression pattern to match the forward slash and leading/trailing spaces
pattern = re.compile(r'^\s*\/\s*|\s*$') 

# Get a list of all files in the folder
files = os.listdir(folder_path)

# Iterate over the files in the folder
for file_name in files:
    if file_name.endswith('.xlsx') or file_name.endswith('.xls'):
        # Apply the regular expression to remove the forward slash and any leading/trailing spaces
        new_file_name = re.sub(pattern, '', file_name)

        # Construct the full paths of the old and new file names
        old_file_path = os.path.join(folder_path, file_name)
        new_file_path = os.path.join(folder_path, new_file_name)
        
        # Rename the file
        os.rename(old_file_path, new_file_path)
        print(f'Renamed "{file_name}" to "{new_file_name}"')
```

### to remove digits and spacing between file extension and word 

```python
import os

# Specify the folder path where the Excel files are located
folder_path = '/Users/apple/Desktop/arunachal pradesh 2019 form 20 xl'

# Recursively iterate over all files and subdirectories within the folder
for root, dirs, files in os.walk(folder_path):
    for file_name in files:
        if file_name.endswith('.xlsx') or file_name.endswith('.xls'):
            # Remove leading digits from the file name
            new_file_name = file_name.lstrip('0123456789. ').strip()

            # Remove extra space before the file extension
            new_file_name = new_file_name.replace(' .', '.')

            # Construct the full paths of the old and new file names
            old_file_path = os.path.join(root, file_name)
            new_file_path = os.path.join(root, new_file_name)

            # Rename the file
            os.rename(old_file_path, new_file_path)
            print(f'Renamed "{file_name}" to "{new_file_name}"')
```

### remove text and spaces from files

```python
import os

# Specify the folder path where the Excel files are located
folder_path = '/Users/apple/Desktop/arunachal pradesh 2019 form 20 xl'

# Specify the texts you want to remove from the file names
texts_to_remove = ['(sc)', '(st)', '(s.c.)','(bl)','(SC)','/','//','-','[st]','(st)']

# Function to recursively rename files in the folder and its subfolders
def rename_files(folder_path, texts_to_remove):
    for root, dirs, files in os.walk(folder_path):
        for file_name in files:
            if file_name.endswith('.xlsx') or file_name.endswith('.xls'):
                # Generate the new file name by replacing the specified texts with an empty string
                new_file_name = file_name
                for text in texts_to_remove:
                    new_file_name = new_file_name.replace(text, '')

                # Construct the full paths of the old and new file names
                old_file_path = os.path.join(root, file_name)
                new_file_path = os.path.join(root, new_file_name)

                # Rename the file
                os.rename(old_file_path, new_file_path)
                print(f'Renamed "{file_name}" to "{new_file_name}"')

# Call the function to rename files recursively
rename_files(folder_path, texts_to_remove)
```

### use it for copying name from one folder to other

```python
import os
import shutil

# Specify the paths of the first and second folders
first_folder_path = '/Users/apple/Desktop/delhi_2019_form_20 _xl'
second_folder_path = '/Users/apple/Desktop/delhi_2014_form_20 _xl'

# Get the list of files in the first folder
first_file_list = os.listdir(first_folder_path)

# Iterate over the files in the first folder
for file_name in first_file_list:
    # Construct the paths for the files in the first and second folders
    first_file_path = os.path.join(first_folder_path, file_name)
    second_file_path = os.path.join(second_folder_path, file_name)

    # Skip directories
    if os.path.isdir(first_file_path):
        continue

    # Check if the file exists in the second folder
    if os.path.exists(second_file_path):
        # Generate a new file name to avoid conflicts
        new_file_name = file_name.split('.')[0] + '_renamed.' + file_name.split('.')[1]
        new_file_path = os.path.join(second_folder_path, new_file_name)

        # Rename the file in the second folder
        os.rename(second_file_path, new_file_path)
    else:
        # Copy the file from the first folder to the second folder
        shutil.copy2(first_file_path, second_folder_path)
```

```python
import os
import pandas as pd
import warnings
import xlrd

warnings.filterwarnings('ignore')

folder_path = "/Users/apple/Desktop/try"  # Replace with the actual folder path

# Iterate through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".xlsx") or filename.endswith(".xls"):  # Process only Excel files
        file_path = os.path.join(folder_path, filename)
        try:
            # Open the Excel file using pandas
            df = pd.read_excel(file_path, header=None)
            
            # Extract the text from merged columns
            cell_value = df.iloc[0, 5:14].fillna('').str.strip().str.cat(sep=' ')

            # Remove any invalid characters from the cell value that are not allowed in filenames
            invalid_chars = r'\/:*?"<>|'
            for char in invalid_chars:
                cell_value = cell_value.replace(char, "")

            # Rename the file using the extracted text
            new_filename = cell_value + ".xlsx"  # You can adjust the file extension if needed

            # Rename the file
            new_file_path = os.path.join(folder_path, new_filename)
            os.rename(file_path, new_file_path)

            print(f"Renamed '{filename}' to '{new_filename}'")
        except Exception as e:
            print(f"Error processing file '{filename}': {str(e)}")
```
