```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
import warnings
warnings.filterwarnings('ignore')

driver = webdriver.Chrome(executable_path="/Users/apple/Downloads/chromedriver3")
url = "https://lgdirectory.gov.in/rptMappedGPNWardforPCAC.do?OWASP_CSRFTOKEN=E0AW-D89D-2BB7-KQ6K-G5T1-4W6D-J2MR-1G2L"
driver.get(url)
driver.maximize_window()
sleep(5)

radio_button = driver.find_element(by=By.XPATH, value='//*[@id="selectAC"]')
radio_button.click()
sleep(3)

# First dropdown
dist1_xpath = '//*[@id="ddSourceState1"]'
driver.find_element(By.XPATH, dist1_xpath).click()
sleep(2)
dist1_options = driver.find_elements(By.XPATH, '//*[@id="ddSourceState1"]/option')


# Click first option of the first dropdown
dist1_options[36].click()  # to change state name change the index name here
sleep(2)



# Second dropdown
dist2_xpath = '//*[@id="parlimentac"]'
driver.find_element(By.XPATH, dist2_xpath).click()
sleep(2)
dist2_options = driver.find_elements(By.XPATH, '//*[@id="parlimentac"]/option')

# Third dropdown
dist3_xpath = '//*[@id="assembly"]'

# Create an empty list to store the options for each option in the second dropdown
options_list = []

# Iterate through options of the second dropdown
for dist2_option in dist2_options:
    # Get the selected option from the second dropdown
    selected_option = dist2_option.text

    # Click the current option of the second dropdown
    dist2_option.click()
    sleep(2)

    # Click the third dropdown
    driver.find_element(By.XPATH, dist3_xpath).click()
    sleep(2)
    
    # Get the options from the third dropdown
    dist3_options = driver.find_elements(By.XPATH, '//*[@id="assembly"]/option')
    
    # Extract all options from the third dropdown
    option_values = [option.text for option in dist3_options]
    
    # Append the selected option and its options to the options_list
    options_list.append([selected_option] + option_values)

    sleep(2)

driver.quit()

# Convert the options_list to DataFrame
df = pd.DataFrame(options_list)

# Save DataFrame to Excel
df.to_excel("options.xlsx", index=False)
```
