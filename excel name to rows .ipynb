```python
import glob
import os
import re
from openpyxl import load_workbook
import xlrd

# Specify the folder path where the Excel files are located
folder_path = "/Users/apple/Desktop/AC election results Latest"

# Get a list of all Excel files in the folder
excel_files = glob.glob(os.path.join(folder_path, "*.xls*"))

# Regular expression pattern to extract the year
pattern = r"\d{4}"  # Matches a four-digit number

# Iterate over each Excel file in the folder
for file_path in excel_files:
    _, file_extension = os.path.splitext(file_path)
    
    # Load the workbook
    if file_extension == '.xlsx':
        workbook = load_workbook(file_path)
    elif file_extension == '.xls':
        workbook = xlrd.open_workbook(file_path, formatting_info=True)
    else:
        continue
    
    # Get the first sheet of the workbook
    if file_extension == '.xlsx':
        sheet = workbook.active
    elif file_extension == '.xls':
        sheet = workbook.sheet_by_index(0)
    else:
        continue
    
    # Extract the year from the file name using regular expressions
    file_name = os.path.basename(file_path)
    match = re.search(pattern, file_name)
    if match:
        year = match.group()
    else:
        year = "Unknown"
    
    # Write the year in the first row (A1 cell)
    if file_extension == '.xlsx':
        sheet["A1"] = year
    elif file_extension == '.xls':
        sheet.write(0, 0, year)
    
    # Save the modified workbook
    if file_extension == '.xlsx':
        workbook.save(file_path)
    elif file_extension == '.xls':
        workbook.save(file_path + "x")  # Save as .xlsx format
```

```python
import os
import re
from openpyxl import load_workbook
import xlrd
import openpyxl

# Specify the folder path where the Excel files are located
folder_path = "/Users/apple/Desktop/AC election results Latest"

# Regular expression pattern to extract the four-digit number
pattern = r"\b\d{4}\b"  # Matches a four-digit number surrounded by word boundaries

# Iterate over each file in the folder
for file_name in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file_name)

    # Check if the file is an Excel file
    if file_name.endswith(".xlsx"):
        # Extract the four-digit number from the file name using regular expressions
        match = re.search(pattern, file_name)
        if match:
            four_digit_number = match.group()
        else:
            four_digit_number = "Unknown"

        # Load the workbook
        workbook = load_workbook(file_path)

        # Get the first sheet of the workbook
        sheet = workbook.active

        # Write the four-digit number in the first row (A1 cell)
        sheet["A1"] = four_digit_number

        # Save the modified workbook
        workbook.save(file_path)

    elif file_name.endswith(".xls"):
        # Extract the four-digit number from the file name using regular expressions
        match = re.search(pattern, file_name)
        if match:
            four_digit_number = match.group()
        else:
            four_digit_number = "Unknown"

        # Read the .xls file using xlrd
        xls_workbook = xlrd.open_workbook(file_path)

        # Create a new .xlsx workbook
        xlsx_workbook = openpyxl.Workbook()

        # Get the first sheet of the .xlsx workbook
        sheet = xlsx_workbook.active

        # Copy data from .xls to .xlsx sheet
        xls_sheet = xls_workbook.sheet_by_index(0)
        for row in range(xls_sheet.nrows):
            for col in range(xls_sheet.ncols):
                sheet.cell(row=row + 1, column=col + 1).value = xls_sheet.cell_value(row, col)

        # Write the four-digit number in the first row (A1 cell)
        sheet["A1"] = four_digit_number

        # Save the modified .xlsx workbook
        xlsx_file_path = os.path.splitext(file_path)[0] + ".xlsx"
        xlsx_workbook.save(xlsx_file_path)

        
```
